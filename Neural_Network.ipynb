Q.1. What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning.

Answer ->

Deep Learning is a subset of Machine Learning (ML) that uses artificial neural networks with multiple layers (hence “deep”) to automatically learn complex patterns and representations from large amounts of data.

**Evolution of Deep Learning :**

- 1950s–1980s: Early Neural Networks
- 1980s–1990s: Backpropagation and Neural Nets Revival
- 2000s: Rise of Machine Learning
- 2010s–Present: Deep Learning Boom

**How Deep Learning Differs from Traditional Machine Learning :**

- It automatically learns features from raw data
- It requires large datasets to perform effectively
- It requires high computational power
- It used in images, audio, video, natural language processing

Q.2. Explain the basic architecture and functioning of a Perceptron. What are its
limitations?

Answer ->

**Architecture of a Perceptron :**

A single-layer perceptron consists of :
- Input Layer
- summation function
- activation function

**Functioning of a Perceptron :**

1. Initialize weights and bias (usually small random values).

2. Feed inputs through the network to compute the output.

3. Compare predicted output y with the actual label t.

4. Update weights using the Perceptron Learning Rule.

5. Repeat for multiple iterations until the model correctly classifies all training samples (or reaches a stopping condition).

**Limitation :**

- Only handles linearly separable data
- It returns binary output only
- No hidden layers
- Sensitive to data scaling
- Slow convergence

Q.4. What is the difference between Loss function and Cost function in neural
networks? Provide examples.

Answer ->

**Difference between Loss funciton and Cost function :**

1. Loss function :     
    - The loss function measures the error for a single training example (or data point).

    - It tells us how far the model’s prediction is from the actual (true) value.

    - It is used in Maen Squared Error.

    - Interpretation : If the model predicts correctly, the loss will be small; if it  predicts poorly, the loss will be large.

2. Cost function :    
    - The cost function is the average of all losses across the entire training dataset (or a batch).

    - It represents the overall performance of the model.

    - It is used for Model parameter optimization.

Q.6. Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.

Answer ->>

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the input and output for AND gate
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0], [0], [0], [1]])  # Expected outputs

# Step 2: Initialize parameters
np.random.seed(42)
weights = np.random.rand(2, 1)      # Random weights for 2 inputs
bias = np.random.rand(1)            # Random bias
learning_rate = 0.1
epochs = 20

# Step 3: Define activation function (Step function)
def activation(x):
    return np.where(x >= 0, 1, 0)

# For tracking error
errors = []

# Step 4: Training loop
for epoch in range(epochs):
    total_error = 0

    for i in range(len(X)):
        # Linear combination
        z = np.dot(X[i], weights) + bias

        # Activation
        y_pred = activation(z)

        # Compute error
        error = y[i] - y_pred
        total_error += int(error != 0)

        # Update weights and bias (Perceptron learning rule)
        weights += learning_rate * error * X[i].reshape(2, 1)
        bias += learning_rate * error

    errors.append(total_error)
    print(f"Epoch {epoch+1}/{epochs} - Total Errors: {total_error}")

# Step 5: Display final weights and bias
print("\nFinal Weights:\n", weights)
print("Final Bias:", bias)

# Step 6: Test the trained model
print("\nTesting AND Gate:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    output = activation(z)
    print(f"Input: {X[i]} → Output: {output[0]}")

# Step 7: Plot error reduction over epochs
plt.plot(range(1, epochs + 1), errors, marker='o')
plt.title("Error Reduction over Epochs (AND Gate)")
plt.xlabel("Epochs")
plt.ylabel("Number of Misclassifications")
plt.grid(True)
plt.show()
```

Q.7.  Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.

Answer ->>

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define input range
x = np.linspace(-10, 10, 200)

# Step 2: Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Step 3: Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Step 4: Create subplots for visualization
plt.figure(figsize=(12, 6))

# Sigmoid
plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, color='blue', linewidth=2)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

# ReLU
plt.subplot(1, 3, 2)
plt.plot(x, y_relu, color='green', linewidth=2)
plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

# Tanh
plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, color='red', linewidth=2)
plt.title("Tanh Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

plt.tight_layout()
plt.show()
```

Q.8. Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.

Answer ->>

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Step 1: Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Step 2: Normalize pixel values (0–255 → 0–1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# Step 3: Convert labels to one-hot encoded vectors
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Step 4: Build the neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),         # Flatten 28x28 image → 784 input neurons
    Dense(128, activation='relu'),         # Hidden layer 1
    Dense(64, activation='relu'),          # Hidden layer 2
    Dense(10, activation='softmax')        # Output layer (10 digits)
])

# Step 5: Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Step 6: Train the model
history = model.fit(X_train, y_train, epochs=5, batch_size=128, verbose=1)

# Step 7: Evaluate on training data
train_loss, train_acc = model.evaluate(X_train, y_train, verbose=0)
print(f"\n✅ Training Accuracy: {train_acc * 100:.2f}%")

# Step 8: Evaluate on test data (optional)
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"✅ Test Accuracy: {test_acc * 100:.2f}%")
```

Q.9.  Visualize the loss and accuracy curves for a neural network model trained
on the Fashion MNIST dataset. Interpret the training behavior.


Answer ->>

```python
# Import libraries
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Step 1: Load the Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Step 2: Normalize the pixel values (0–255 → 0–1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# Step 3: Convert labels to one-hot encoded format
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Step 4: Build the neural network model
model = Sequential([
    Flatten(input_shape=(28, 28)),     # Input layer
    Dense(128, activation='relu'),     # Hidden layer 1
    Dense(64, activation='relu'),      # Hidden layer 2
    Dense(10, activation='softmax')    # Output layer
])

# Step 5: Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Step 6: Train the model
history = model.fit(X_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_data=(X_test, y_test),
                    verbose=1)

# Step 7: Plot training & validation accuracy/loss curves
plt.figure(figsize=(12, 5))

# Accuracy Curve
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Loss Curve
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
